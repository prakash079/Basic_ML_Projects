```python
#but first let's do some changes here we can change max_depth
#that will lead to a good accuracy
```

```python
for i in range(1,10):
    dt=DecisionTreeClassifier(max_depth=i)
    dt.fit(X_train,Y_train)
    prediction=dt.predict(X_test)
    accuracy_dt=accuracy_score(Y_test,prediction)
    print('for '+str(i)+' accuracy is '+str(accuracy_dt*100))
```

```python
#so by this we can see that for max_depth=1 accuracy is greatest
```
